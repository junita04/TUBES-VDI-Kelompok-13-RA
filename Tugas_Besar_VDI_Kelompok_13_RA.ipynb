```python
# Mengonversi kolom 'utcTimestamp' menjadi datetime
df['utcTimestamp'] = pd.to_datetime(df['utcTimestamp'], format='%a %b %d %H:%M:%S +0000 %Y')

# Mengurutkan berdasarkan userId dan utcTimestamp untuk memastikan data terurut dengan benar
df = df.sort_values(by=['userId', 'utcTimestamp'])

# Menghitung selisih waktu antara dua baris berturut-turut
df['timeSpent'] = df.groupby('userId')['utcTimestamp'].diff().dt.total_seconds()

# Menampilkan hasil
print(df)
```

```python
# menghapus nilai NaN
df = df.dropna()
df
```

```python
# Menghapus kolom yang tidak diperlukan
df = df.drop(columns=['venueId', 'venueCategoryId', 'timezoneOffset'])

# Menampilkan dataset setelah preprocessing
print(df)
```

**Kategori tempat paling sering dikunjungi**

```python
# Menghitung total timeSpent per venueCategory dan mengurutkannya
category_time_spent = df.groupby('venueCategory')['timeSpent'].sum().sort_values(ascending=False)
```

```python
# Memilih 10 kategori tempat dengan waktu terbanyak yang dihabiskan
top_categories = category_time_spent.nlargest(10)
```

```python
# Visualisasi dengan bar chart
plt.figure(figsize=(10, 6))
top_categories.plot(kind='bar', color='skyblue')
plt.title('10 Kategori Tempat Teratas dengan Waktu Terbanyak yang Dihabiskan', fontsize=14)
plt.xlabel('Venue Category', fontsize=12)
plt.ylabel('Total Time Spent (seconds)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**waktu kunjungan paling sibuk untuk kategori tempat paling sering dikunjungi**

```python
# Filter data untuk kategori
venue_filter = df[df['venueCategory'].isin(['Train Station', 'Subway'])]
```

```python
# Extract jam dari kolom 'utcTimestamp'
venue_filter['hour'] = venue_filter['utcTimestamp'].dt.hour
```

```python
# Menghitung jumlah kunjungan per jam
visits_per_hour = venue_filter.groupby('hour').size()
```

```python
# Visualisasi jumlah kunjungan per jam
plt.figure(figsize=(10, 6))
visits_per_hour.plot(kind='bar', color='lightcoral')
plt.title('Jam Paling Sibuk untuk Kunjungan Train Station dan Subway', fontsize=14)
plt.xlabel('Jam', fontsize=12)
plt.ylabel('Jumlah Kunjungan', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Kategori tempat yang menunjukkan perbedaan paling besar dalam frekuensi kunjungan antara hari kerja dan akhir pekan di Tokyo

```python
# Menambahkan kolom 'dayofweek' untuk mengetahui hari dalam minggu (0=Senin, 6=Minggu)
df['dayofweek'] = df['utcTimestamp'].dt.dayofweek
```

```python
# Membuat kolom 'weekend' untuk memisahkan hari kerja (0-4) dan akhir pekan (5-6)
df['isWeekend'] = df['dayofweek'].apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')
```

```python
# Menghitung frekuensi kunjungan per venueCategory dan jenis hari (weekend atau weekday)
category_frequency = df.groupby(['venueCategory', 'isWeekend']).size().unstack(fill_value=0)
```

```python
# Menghitung perbedaan frekuensi kunjungan antara weekday dan weekend
category_frequency['Difference'] = category_frequency['Weekend'] - category_frequency['Weekday']
```

```python
# Mengurutkan kategori tempat berdasarkan perbedaan terbesar
top_categories = category_frequency['Difference'].abs().nlargest(10).index
```

```python
# Visualisasi: bar plot untuk 10 kategori tempat dengan perbedaan terbesar antara weekday dan weekend
top_category_data = category_frequency.loc[top_categories]
```

```python
top_category_data[['Weekday', 'Weekend']].plot(kind='bar', figsize=(12, 6), color=['skyblue', 'lightcoral'])
plt.title('Perbandingan Frekuensi Kunjungan Hari Kerja vs Akhir Pekan untuk 10 Kategori Teratas di Tokyo', fontsize=14)
plt.xlabel('Kategori Tempat', fontsize=12)
plt.ylabel('Jumlah Kunjungan', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(title='Jenis Hari', loc='upper right')
plt.tight_layout()
plt.show()
```
